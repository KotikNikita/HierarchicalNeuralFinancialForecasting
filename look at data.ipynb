```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df_daily = pd.read_pickle('data/NASDAQ_hist_daily_06102021-091329.pkl')
df_hour = pd.read_pickle('data/NASDAQ_hist_intra_gran_60min_18092021-104113.pkl')
df_5min = pd.read_pickle('data/NASDAQ_hist_intra_gran_5min_18092021-163914.pkl')
df_1min = pd.read_pickle('data/NASDAQ_hist_intra_gran_1min_18092021-193247.pkl')
```

```python
start = dt.datetime(2019,10,1,5,0,0)
end = dt.datetime(2019,10,2,23,0,0)

plt.figure(figsize=(10, 7))
plt.grid()

df_nflx = df_daily[(df_daily['Ticker'] == 'NFLX') & (df_daily['Time'] >= start) & (df_daily['Time'] <= end)]
plt.scatter(df_nflx['Time'], df_nflx['Close'], label='daily')

df_nflx = df_hour[(df_hour['ticker'] == 'NFLX') & (df_hour['time'] >= start) & (df_hour['time'] <= end)]
plt.scatter(df_nflx['time'], df_nflx['close'], label='hour')

df_nflx = df_5min[(df_5min['ticker'] == 'NFLX') & (df_5min['time'] >= start) & (df_5min['time'] <= end)]
plt.scatter(df_nflx['time'], df_nflx['close'], label='5 min', s=5)

df_nflx = df_1min[(df_1min['ticker'] == 'NFLX') & (df_1min['time'] >= start) & (df_1min['time'] <= end)]
plt.scatter(df_nflx['time'], df_nflx['close'], label='1 min', s=1)

plt.legend()
```
